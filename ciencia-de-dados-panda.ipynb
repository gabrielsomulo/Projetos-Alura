```python
imoveis = ['Apartamento',
       'Casa de Condomínio',
       'Loja Shopping/ Ct Comercial', 'Chácara', 'Loteamento/Condomínio',
       'Sítio', 'Pousada/Chalé', 'Studio', 'Hotel', 'Indústria']
```

```python
dados.query('@imoveis in Tipo')
```

```python
df_query_not = dados.query('@imoveis not in Tipo')
```

```python
df_preco_tipo2 = df_query_not.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
pd.read_csv(url2)
```

```python
dados2 = pd.read_csv(url2, sep=';')

dados2
```

```python
dados2.groupby('Tipo')[['Quartos','Vagas']].mean().sort_values('Quartos')
```

```python
bairros = dados2.Tipo.value_counts(normalize=True)

bairros.plot(kind='barh', figsize=(14, 10), color='green', xlabel = 'Percentual', ylabel = 'Tipos')
```

```python
bairros_caros = dados2.groupby('Tipo')[['Valor']].mean().sort_values('Valor').tail()
```

```python
bairros_caros
```

```python
bairros_caros.plot(kind='bar', figsize=(10, 10), color='blue', xlabel = 'Tipos', ylabel = 'Valor')
```
