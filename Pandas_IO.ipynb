## APIs JSON

```python
import requests
import json
```

```python
dados_usuarios_cru = requests.get('https://jsonplaceholder.typicode.com/users')
```

```python
resultado = json.loads(dados_usuarios_cru.text)
```

Ao usar o argumento sep=' _ ' no método pd.json_normalize, você garante que os subcampos de dados aninhados sejam desaninhados com os níveis unidos por um sublinhado ( _ ). Isso facilita o acesso e a manipulação dos dados no DataFrame, mantendo uma estrutura mais clara.

```python
dados_usuarios_normalizado = pd.json_normalize(resultado, sep='_')
```

```python
dados_usuarios_normalizado
```
