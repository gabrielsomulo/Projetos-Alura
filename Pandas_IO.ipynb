## Importação seletiva

```python
dados_linhas_csv1 = pd.read_csv(url_csv1, nrows=5)
dados_linhas_csv1
```

```python
dados_colunas_csv1 = pd.read_csv(url_csv1, nrows=3, usecols=['Id', 'Year_Birth','Income'])
dados_colunas_csv1
```

```python
dados_colunas_index_csv1 = pd.read_csv(url_csv1, nrows=3, usecols=[0, 3,5])
dados_colunas_index_csv1
```

## Gravação de iformações

```python
dados_colunas_index_csv1.to_csv('dados_colunas_index_csv1.csv')
```

```python
# Sem tratamento de dados
coleta_de_dados_com_index = pd.read_csv('dados_colunas_index_csv1.csv')
coleta_de_dados_com_index
```

```python
# Com tratamento de dados
dados_colunas_index_csv1.to_csv('dados_colunas_index_csv1.csv', index=False)
coleta_de_dados_sem_index = pd.read_csv('dados_colunas_index_csv1.csv')
coleta_de_dados_sem_index
```
